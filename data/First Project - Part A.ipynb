6. Check if there are any missing values.

```python
df.isnull().values.any()
```

Explanation: 
- df.isnull() will first return a dataframe with True if the value of the record is null and False if is not null.
- df.isnull().values puts the records values of the dataframe into an array (True, False).
- df.isnull().values.any() Look into the array and returns True if there are any True values otherwise returns False.

The Result returns True. It means that we have some missing values
