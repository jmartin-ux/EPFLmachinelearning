C. Bar plot of the Happiness Score

1. Obtain a horizontal bar plot of the Happiness Score of the top 10 happiest countries. Your bar chart should have the names of the countries are listed vertically along the 
y-axis and the x-axis should have labels for each number from 0 to 8. Ensure that the chart has an appropriate title and labels.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df2 = df[['Country','Happiness Score']].sort_values('Happiness Score',ascending=False)
```

```python
df2.set_index(np.arange(len(df2.index)),inplace=True)
```

```python
fig, ax = plt.subplots()
countries = df2[0:10]['Country']
happiness = df2[0:10]['Happiness Score']
colormap = plt.cm.Paired(range(0, len(countries)))
ax.barh(countries, happiness,color=colormap, edgecolor='black')
ax.set_title('The 10 happiest countries in the World')
ax.set_ylabel('Country')
ax.set_xlabel('Happiness Score')
plt.xticks(np.arange(9))
plt.gca().invert_yaxis()
plt.show()
```

Explanaition
- First I create a figure and an axis
- I select the first 10 countries from the dataframe of the countries sorted by the Happiness score and put into a Series that I will use for y axis
- I select the first 10 Happiness Scores from the dataframe of the countries sorted by Happiness score and put into a Series that I will use for the x axis
- I define a different color for each country of the variable countries by using the Matplotlib built-in function matplotlib.cm and using the Qualitive colormap Paired to better visualize the difference between each other
- I then create an horizontal bar with content of the Countries Series as x axis, the hapiness Series as y axis, colormap array as color of the bars and with edgecolor black for the bars
- I set x and y labels with the values of the Happiness score and Country Series
- I set the x ticks as an array from 0 to 8 to comply the requirement
- I invert the axis to sort descending by the highest happiness score

G. Probabilities
1. Compute the probability that randomly selected country with Happiness score over 6.0 is from Western Europe. You will have to use pandas to count the appropriate quantities.

```python
import random
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df_happiness_score=df[df["Happiness Score"] > 6.0][['Country','Happiness Score','Region']]
experiments = 10000
proper_region = 0

for i in range(experiments):
    df_random_happiness=df_happiness_score.iloc[random.randrange(0,len(df_happiness_score),1)]
    if df_random_happiness['Region'] == 'Western Europe':
        proper_region += 1
proper_region / experiments
```

Explaination:
- I define the Series df_happiness_score with the records of the dataframe with hapiness score higher than 6.0 and displaying only the columns Country, Happiness Score and Region
- I loop then the experiments until 10000
- In the loop I set the Series df_random_happiness to the record selected randomly from the dataset df_happiness_score
- If the Series selected randomly is equal to 'Western Europe' then I increase my counter proper_region by 1
- At the end of the loop I divide the result of my counter proper_region my the number of experiments and I get 0.3864.

```python
rows=df['Country']
columns=df['Region'].unique()
```

```python
a = np.zeros((153,7))
n_countries=len(rows)
n_regions=len(columns)

a = np.zeros((n_countries,n_regions))

for i in range(n_countries):
    for j in range(n_regions):
        if (df[(df['Country'] == rows[i]) & (df['Region'] == columns[j])].any(axis=None)):
            a[i][j] = "1"
        else:
            a[i][j] = "0"
a
```

Explaination:
- I defined two Series, one with the content of the values of the column Country and the other with the column Region (by selecting unique in order to get only one occurency per each different Region).
- Then I inizialize my matrix a by setting all records to zero
- I loop then by the countries with index i and by the regions with index j and make if condition:
    - if combination of the values in the arrays row[i] and columns[j] exists in the dataframe df then set the corresponding value of the matrix to 1 otherwise set it to 0
